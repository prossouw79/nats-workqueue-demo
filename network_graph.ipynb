```python
from pyvis.network import Network
import json, os, shutil, datetime
from html2image import Html2Image

# hti = Html2Image(size=(800, 800))

with open("results/collection.json") as f:
    data = json.load(f)

    for i in range(len(data)):
        item = data[i]
        print(str(i + 1) +"/" + str(len(data)))
        uid = item['uid']
        stops = item['stops']

        # print(stops)

        net = Network()

        node_index = 1
        hosts_uniq = []

        last_time = None
        current_time = None

        for s in range(len(stops)):
            stop = stops[s]
            host = stop['host']

            is_first = s == 0
            is_last = s == (len(stops) - 1)

            ts = datetime.datetime.strptime(stop['timestamp'], '%Y-%m-%dT%H:%M:%S.%fZ')
            ms_passed=0

            if current_time is not None:
                last_time = current_time
                current_time = ts

            current_time = ts

            if current_time is not None and last_time is not None:
                diff = current_time - last_time
                ms_passed = round(diff.total_seconds() * 1000)

            c="#5e5e5e"

            if is_first:
                c="#3da831"

            if is_last:
                c="#ff1900"

            if host not in hosts_uniq:
                net.add_node(host, color=c)
                hosts_uniq.append(host)
            if s > 0:
                lasthost = stops[s-1]['host']
                link_title=str(ms_passed)+"ms"
                net.add_edge(host, lasthost, value=1, title=link_title)

        html_path = "graphs/" + str(i) + '.html'
        net.save_graph(html_path)

        # image_filename=uid + ".png"
        # hti.screenshot(html_file=html_path,save_as=image_filename)
        # os.remove(html_path)
        # shutil.move(image_filename, "graphs/" + image_filename)
```
